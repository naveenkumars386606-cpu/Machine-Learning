```python
# Label Encoding
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
for i in df.columns:
  if(df[i].dtype == "object"):
    df[i] = le.fit_transform(df[i])
```
