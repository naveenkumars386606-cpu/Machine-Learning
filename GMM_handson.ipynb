```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn.cluster import KMeans
```

```python
from sklearn.datasets import make_blobs
```

```python
np.random.seed(42)
```

```python
x,y=make_blobs(n_samples=500,centers=3,cluster_std=[1.0,2.5,0.5],random_state=42)
```

Parameter Explanation
n_samples=500

Creates:

500 data points
centers=3

Creates:

3 clusters

Something like:

Cluster 0
Cluster 1
Cluster 2
cluster_std=[1.0, 2.5, 0.5]

This is the most interesting parameter.

It controls the spread (standard deviation) of each cluster.

Cluster 0 → std = 1.0
Cluster 1 → std = 2.5
Cluster 2 → std = 0.5

Visualization:

Cluster 0
   Medium spread

Cluster 1
   Very wide spread

Cluster 2
   Very compact

```python
GMM = GaussianMixture(n_components=3,random_state=42)
gmm_labels=GMM.fit_predict(x)
```

```python
gmm_labels
```

```python
Kmeans=KMeans(n_clusters=3,random_state=42)
kmeans_labels=Kmeans.fit_predict(x)
```

```python
kmeans_labels
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharex=True, sharey=True)

axes[0].scatter(x[:, 0], x[:, 1], c=y, cmap='viridis', s=30)
axes[0].set_title("True Clusters")

axes[1].scatter(x[:, 0], x[:, 1], c=gmm_labels, cmap='viridis', s=30)
axes[1].set_title("GMM Clusters")

axes[2].scatter(x[:, 0], x[:, 1], c=kmeans_labels, cmap='viridis', s=30)
axes[2].set_title("K-Means Clusters")

plt.tight_layout()
plt.show()
```

```python
A = np.array([
    [5,4,0],
    [4,5,2],
    [0,1,5]
])
```

```python
U,S,VT = np.linalg.svd(A)
```

```python
print(U)
```

```python
print(S)
```

```python
print(VT)
```

```python
#K = 2
k = 2
U_k = U[:, :k]
S_K = np.diag(S[:k])
VT_K = VT[:k,:]
```

```python
NewA = np.dot(U_k,np.dot(S_K,VT_K))
```
